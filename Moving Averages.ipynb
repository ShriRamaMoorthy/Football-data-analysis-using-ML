```python
# use pandas to load and manipulate the data, matplotlib to plot
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tottenham prem xg.csv')
```

```python
df.head()
```

```python
df=df.drop(['Attendance','Notes'],axis=1)
```

```python
df.head()
```

```python
df['match'] = range(1,39)
```

```python
df['xgSMA'] = df['xG'].rolling(window=3).mean()
df['xgaSMA'] = df['xGA'].rolling(window=3).mean()
```

```python
plt.plot(df.match,df.xgSMA,color='blue')
plt.plot(df.match,df.xgaSMA,color='red')

plt.xlabel("Matchdays")
plt.ylabel("xG - Expected goals")

plt.legend(['Tottenham Hotspur F.C','Rest of the PL Teams'])
```
